```python
import pandas as pd
marks_science=[78,56,89,92,45,62]
marks_maths=[78,56,89,92,45,62]
grade=['b','c','a','a','d','b']
data=pd.DataFrame({
    "science marks":marks_science,
    "Maths marks":marks_maths,
    "Total Grade":grade
})
```

```python
#option 1
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
le.fit(data["Total Grade"])
data["Total Grade"]=le.transform(data['Total Grade'])
data
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('btissue.csv')
df
```

```python
pred=df.iloc[:,0:9]
pred
```

```python
tar=df.iloc[:,9]
tar
```

```python
from sklearn.model_selection import train_test_split
```

```python
pred_train,pred_test,tar_train,tar_test=train_test_split(pred,tar,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(random_state=42)
```

```python
rd=model.fit(pred_train,tar_train)
```

```python
predict=model.predict(pred_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(tar_test,predict,normalize=True)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(tar_test,predict)
```

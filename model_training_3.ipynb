```python
import numpy as np
import pandas as pd
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import v_measure_score
```

```python
data=pd.read_csv("Dataset_spine.csv")
```

```python
data
```

```python
data_woc=data.iloc[:,0:12]
data_woc
```

```python
data_class=data.iloc[:,12]
data_class
```

```python
f1=data_woc['Col1'].values
f1
```

```python
f2=data_woc['Col5'].values
f2
```

```python
f3=data_woc['Col9'].values
f3
```

```python
x=np.array(list(zip(f1,f2,f3)))
```

```python
x
```

```python
kmean=KMeans(n_clusters=2,random_state=42)
```

```python
model=kmean.fit(x)
```

```python
cluster_labels=kmean.predict(x)
```

```python
v=v_measure_score(cluster_labels,data_class)
```

```python
v
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil=silhouette_score(x,cluster_labels,metric='euclidean',sample_size=len(data))
```

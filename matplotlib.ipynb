```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,color="red",linestyle="--",marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")
plt.grid()
plt.show()
```

```python
x=range (-15,16)
y=[x**2 for x in x]
plt.plot(x,y,color="red",linestyle="-.",marker="o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")
plt.grid()
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,'r',label="sin(x)")
plt.plot(x,y2,'g',label="cos(x)")
plt.legend(loc=9)
plt.show()
```

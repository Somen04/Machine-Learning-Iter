```python
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,'r',label="sin(x)")
plt.plot(x,y2,'g',label="cos(x)")
plt.legend(loc=9)
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
plt.scatter(x,y1)
plt.scatter(x,y2)
```

```python

np.random.seed(40) # For reproducibility, the same random numbers every time you run the code.42:seed value
X = np.random.rand(100) # 50 random numbers between 0 and 1
Y = np.random.rand(100)
plt.scatter(X, Y, color='red', marker='o', s=100, edgecolors='black', alpha=0.5, label='Scatter Plot')
#s-> Sets the marker size.
#edgecolors-> Sets the color of the marker edges.
#alpha-> Adjusts the transparency of the markers.
plt.legend() #position the legend on your plot. Location Code 9-upper center
plt.show()
```

```python
# Sample data
data = [10, 15, 20, 25, 30]
df=pd.DataFrame(data)
print(df.describe())
# Create a box plot
plt.boxplot(data)

# Add title and labels
plt.title("Box Plot Example")
plt.ylabel("Values")

plt.show()
```

```python
# Histogram Chart/Plot:
data = [12, 15, 14, 10, 18, 20, 25, 30, 28, 22, 15, 19]

# Create a histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.title("Histogram of Data")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
'''To keep improving their services, XYZ group of hotels have asked all the three hotels 
to get feedback form filled by their customers at the time of checkout. 
After getting ratings on a scale of (1–5) on factors such as Food, Service, Ambience, 
Activities, Distance from tourist spots they calculate the average rating and store it in a CSV file.'''

import pandas as pd
# Create a sample dictionary of 
data = { 'Year': [2014, 2015, 2016, 2017, 2018], 
             'Sunny Bunny Resort': [4.75, 2.5, 3.5, 4, 1.5], 
             'Happy Lucky Resort': [3, 4 , 2.5 , 2 , 4.5], 
             'Breesy Windy Resort': [4.75, 2 , 3 , 3.5 , 1]}
# Convert dictionary to DataFrame
df1 = pd.DataFrame(data)
# Save DataFrame to CSV file
df1.to_csv('Data.csv', index=False)
print("CSV file 'Data.csv' created successfully!")
data1 = pd.read_csv('Data.csv')
df2= pd.DataFrame(data1)
#plot a box plot for the DataFrame 'df' with a title
df2.drop('Year', axis=1).plot(kind='box',title='Compare Resorts')
#set xlabel,ylabel
plt.xlabel('Resorts')
plt.ylabel('Rating (5 years)')
#display the plot
print(df2)
plt.show()
```
